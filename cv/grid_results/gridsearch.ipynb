```python
import joblib
import profile_default
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
cv_results = joblib.load('results.joblib')
```

```python
# cv results
cv_results = pd.DataFrame(cv_results)
cv_results
```

```python
cv_results[cv_results['param_kernel']=='rbf'].iloc[:,8:]
```

```python
# converting C to int type
cv_results['param_C'] = cv_results['param_C'].astype('int')  

plt.figure(figsize=(16,8))

# subplot 1
plt.subplot(121)
gamma_01 = cv_results[cv_results['param_gamma']==2]

plt.plot(gamma_01["param_C"], gamma_01["mean_test_score"])
plt.plot(gamma_01["param_C"], gamma_01["mean_train_score"])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title("Gamma=2")
plt.ylim([0.00, 1.02])
plt.legend(['test accuracy', 'train accuracy'], loc='upper left')
plt.xscale('log')

# subplot 2/3
plt.subplot(122)
gamma_001 = cv_results[cv_results['param_gamma']==0.5]

plt.plot(gamma_001["param_C"], gamma_001["mean_test_score"])
plt.plot(gamma_001["param_C"], gamma_001["mean_train_score"])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title("Gamma=0.5")
plt.ylim([0.00, 1.02])
plt.legend(['test accuracy', 'train accuracy'], loc='upper left')
plt.xscale('log')

plt.savefig('GridSearch.png')
```
